# Data Preprocessing

<font color="#C63A7A" size=3><b>REPLACE ALL MISSING VALUES BY NAN </b></font> 

```python
data=data.replace('?',np.nan)
```

<font color="#C63A7A" size=3><b>REPLACE ALL MISSING VALUE BY THE MAJORITY VALUE FOR EACH FEATURE </b></font> 

```python
data=pd.read_csv(r"census.csv")
majorityWorkclass=data["workclass"].mode()[0]
majorityOccupation=data["occupation"].mode()[0]
majorityNativeCcountry =data["native-country"].mode()[0]
print(f"majority value for the column workclass is {majorityWorkclass}")
print(f"majority value for the column occupation is {majorityOccupation}")
print(f"majority value for the column native-country is {majorityNativeCcountry}")
```

```python
data=data.replace('?',np.nan)
```

```python
data['workclass']=data['workclass'].fillna(majorityWorkclass) 
data['occupation']=data['occupation'].fillna(majorityOccupation) 
data['native-country']=data['native-country'].fillna(majorityNativeCcountry) 
```

```python
data.isnull().sum()
```

<font color="#C63A7A" size=3><b>CONVERT THE VALUE OF INCOME TO 0 FOR '<=50K' AND TO 1 FOR ' >50K' </b></font> 

```python
data.income=data.income.replace('<=50K','0')
data.income=data.income.replace('>50K','1')
data.head(4)
```

<font color="#C63A7A" size=3><b>CHANGE THE NAME OF SOME COLUMNS TO BE FIT WITH PYTHON </b></font> 

```python
data=data.rename(columns={'marital-status':'marital_status',
                          'education-num':'education_num',
                          'capital-gain':'capital_gain',
                          'capital-loss':'capital_loss',
                          'hours-per-week':'hours_per_week',
                          'native-country':'native_country'})
data.head(3)
```

<font color="#C63A7A" size=3><b>CONVERT ALL DATA TO NUMERIC VALUE </b></font> 

```python
from sklearn.preprocessing import LabelEncoder
workclass=LabelEncoder()
education=LabelEncoder()
marital_status=LabelEncoder()
occupation=LabelEncoder()
relationship=LabelEncoder()
race=LabelEncoder()
sex=LabelEncoder()
native_country=LabelEncoder()
data['workclass']=workclass.fit_transform(data['workclass'])
data['education']=education.fit_transform(data['education'])
data['marital_status']=marital_status.fit_transform(data['marital_status'])
data['occupation']=occupation.fit_transform(data['occupation'])
data['relationship']=relationship.fit_transform(data['relationship'])
data['race']=race.fit_transform(data['race'])
data['sex']=sex.fit_transform(data['sex'])
data['native_country']=native_country.fit_transform(data['native_country'])
data.head(10)
```

<font color="#C63A7A" size=3><b>SPLIT THE DATA 30% FOR TEST AND 70% FOR TAINING </b></font>

```python
from sklearn.model_selection import train_test_split
#drop the class income
X=data.drop('income',axis=1)
#put the row class income alone
y=data[['income']]
#split the data
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=10)
```

```python
from sklearn.feature_selection import VarianceThreshold

selector = VarianceThreshold(threshold = 0.025)#methode de selection des attibute 0.025 seuil bech yna7i colonne
selector.fit(X_train, y)
#selector.fit_transform(X_train_enc)
x_reduced_train=selector.transform(X_train)#transformer 
x_reduced_test=selector.transform(X_test)
selector.variances_.tolist()
```

```python
print(f"the number of instances for training are {len(X_train)}   70%")
print(f"the number of instances for testing are {len(X_test)}     30%")
print(y)
```
